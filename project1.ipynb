```python
df1=df[['iyear','imonth','iday','country_txt','region_txt','city','latitude','longitude','success','attacktype1_txt','targtype1_txt','natlty1_txt','gname','motive','weaptype1_txt','nkill','nwound']]
```

```python
df1.rename(columns={'iyear':'year','imonth':'month','iday':'day','country_txt':'country','region_txt':'region','attacktype1_txt':'attacktype','nwound':'wound','targtype1_txt':'targettype','natlty1_txt':'nationality','gname':'group','weaptype1_txt':'weapon','nkill':'kill'},inplace=True)
```

```python
df1.isnull().sum()
```

 **Visualisation**

```python
count=df1['country'].value_counts()[:10]
count.plot(kind='bar')
plt.title("Top 10 terror attack victim countries")
plt.xlabel("Countries")
plt.ylabel("No. of Attacks")
```

```python
x_year = df1['year'].unique()
y_count_years = df1['year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (20,10))
sns.barplot(x = x_year,
           y = y_count_years)
plt.xticks(rotation = 90)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
```

```python
count=df1['region'].value_counts()
count.plot(kind='bar')
plt.title("Terror attacks region wise")
plt.xlabel("Regions")
plt.ylabel("No. of Attacks")
```

```python
count=df1['success'].value_counts()
count.plot(kind='bar')
plt.title("Success or Failure of terror attack")
plt.xlabel("Success = 1 , Failure = 0")
plt.ylabel("No. of Attacks")
```

```python
x_attacktype = df1['attacktype'].unique()
y_count_attacktype = df1['attacktype'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (20,10))
sns.barplot(x = x_attacktype,
           y = y_count_attacktype)
plt.xticks(rotation = 90)
plt.xlabel('Type of Attack')
plt.ylabel('Count of attack type')
plt.title('Different type of attack')
plt.show()
```

```python
x_target = df1['targettype'].unique()
y_count_target = df1['targettype'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (20,10))
sns.barplot(x = x_target,
           y = y_count_target)
plt.xticks(rotation = 90)
plt.xlabel('Types of targets')
plt.ylabel('No. of targets')
plt.title('Different types of target')
plt.show()
```

```python
count=df1['nationality'].value_counts()[:20]
count.plot(kind='bar')
plt.title("Nationality of Attacker")
plt.xlabel("Country Attacker belongs")
plt.ylabel("count of Nationality")
```

```python
x_group = df1['group'].unique()[:20]
y_count_group = df1['group'].value_counts(dropna = False).sort_index()[:20]
plt.figure(figsize = (20,10))
sns.barplot(x = x_group,
           y = y_count_group)
plt.xticks(rotation = 90)
plt.xlabel('Terror Groups Name')
plt.ylabel('Count of terror attack by terror groups')
plt.title('Terror Groups')
plt.show()
```

```python
count=df1['kill'].value_counts()[:10]
count.plot(kind='bar')
plt.title("kills in an attack")
plt.xlabel("No. of death in a attack")
plt.ylabel("Count of Attack")
```
